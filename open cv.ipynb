```python
Image = cv2.imread(r"C:\Users\ELCOT\Pictures\cute rabbit photo.jpg")

gray = cv2.cvtColor(Image, cv2.COLOR_BGR2GRAY)
hsv = cv2.cvtColor(Image, cv2.COLOR_BGR2HSV)
lab = cv2.cvtColor(Image, cv2.COLOR_BGR2LAB)

cv2.imshow("Grayscale", gray)
cv2.imshow("HSV", hsv)
cv2.imshow("LAB", lab)
cv2.waitKey(0)# cv2.waitkey(delay) waits for a key press.
# delay:delay in milliseco
#0 means wait indefinitely
cv2.destroyAllWindows()
```

```python
#Resizing with differents methods:

Image = cv2.imread(r"C:\Users\ELCOT\Pictures\cute rabbit photo.jpg")

#Resize with inter_linear
resized_linear= cv2.resize(Image, (400,300),
interpolation= cv2.INTER_LINEAR)

#Resize with inter_cubic
resized_cubic= cv2.resize(Image, (400,600), 
interpolation= cv2.INTER_CUBIC)
cv2.imshow("Resized (linear)",resized_linear)
cv2.imshow("Resized (cubic)", resized_cubic)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#crop a region of interest (ROI)

Image = cv2.imread(r"C:\Users\ELCOT\Pictures\cute rabbit photo.jpg")

# define crop area (top-left: x=100,y=50; bottom-rigth:x=400,y=300)

cropped= Image[50:300,100:400]
cv2.imshow("cropped image", cropped)
cv2.waitKey(0)
cv2.destroyAllWindows()

```
